```python

import json
import matplotlib.pyplot as plt

# Load the JSON file
with open('noindefnns_train_stats.json', 'r') as f:
    data = json.load(f)

# Extract log history entries
log_history = data.get('log_history', [])[:-1]

# Get epoch and loss values from the log history
steps = [entry['step'] for entry in log_history if 'step' in entry]
losses = [entry['loss'] for entry in log_history if 'loss' in entry]

# Plot loss vs epoch
plt.figure(figsize=(10, 6))
plt.plot(steps, losses, marker='o', markersize=3, linestyle='-', label='Training Loss')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.title('No AANN + No Indefinite NNS Loss vs Steps')
plt.legend()
plt.grid(True)
plt.show()
```
